```python
import pandas as pd

# Αντικατέστησε με το πραγματικό όνομα του αρχείου σου
df = pd.read_csv('/content/smoking_bias_metrics.csv')

# Εμφάνισε τις πρώτες γραμμές για να δεις τις στήλες
df.head()
```

```python
min_val = df['Disparate Impact'].min()
max_val = df['Disparate Impact'].max()
range_val = max_val - min_val

print(f'Ελάχιστο: {min_val}')
print(f'Μέγιστο: {max_val}')
print(f'Εύρος τιμών: {range_val}')
```
